```python
a=eval(input("Enter the value of a:"))
b=eval(input("Enter the value of b:"))
def add():
    print(a+b)
    
def sub():
    print(a-b)
    
def mul():
    print(a*b)
    
def div():
    print(a/b)

def calculator():
    option=eval(input("Enter the number between 1-4:"))
    if option==1:
        add()
    elif option==2:
        sub()
    elif option==3:
        mul()
    elif option==4:
        div()

calculator()
```

```python
def add(a,b):
    print(a+b)
    
def sub(a,b):
    print(a-b)
    
def mul(a,b):
    print(a*b)
    
def div(a,b):
    print(a/b)

def calculator():
    option=eval(input("Enter the number between 1-4:"))
    a=eval(input("Enter the value of a:"))
    b=eval(input("Enter the value of b:"))
    if option==1:
        add(a,b)
    elif option==2:
        sub(a,b)
    elif option==3:
        mul(a,b)
    elif option==4:
        div(a,b)

calculator()
```
